```r
library(tidyr)
library(ggplot2)
```

```r
df = read.csv("NetflixViewingHistory.csv")
df
```

```r
df1 <- df %>%
  separate(col = Title, into = c("title", "season", "episode"), sep = ': ') %>% drop_na()
```

```r
netflix_day <- df1 %>%
  count(Date)
```

```r
ggplot(aes(x = netflix_day$Date, y = netflix_day$n), data = netflix_day) +
  geom_col(color = c("red")) +
  theme_minimal() +
  ggtitle("Episode Watched Per Day") +
  labs(x = "Date", y = "Episode Wachted") 
```

```r
df2 <- df %>%
  separate(col = Date, into = c("month", "day", "year"), sep = '/')
df2 <- df2 %>% count(month)
df2 <- df2[order(df2$month),]
df2 <- df2[order(df2$n),]
pie(df2$n, df2$month)
title("Episode Wached by Month")
```
