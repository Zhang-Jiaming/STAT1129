Q5

```python
marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}
```

```python
x = np.array(list(marks.keys()))
y = np.array(list(marks.values()))

plt.title("Student's Grade")
plt.bar(x, y)
plt.show()
```

```python
plt.pie(y, labels = x)
plt.legend(bbox_to_anchor=(1.05,1.0))
plt.title("Student's Grade")
plt.show()
```
