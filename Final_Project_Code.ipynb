```python
def remove_format(word):
    #create a string of characters need to be removed
    format = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~\n"
    #convert to lowercase
    word = word.lower()
    #remove unwanted characters
    for char in format:
        word = word.replace(char, "")
    return word

def check_exist_in_dict(word_dict, word,line_number):
    #avoid double counting line number for same word
    for existing_value in word_dict.get(word):
        #if line_number already exist in dictionary, break and return True
        if existing_value == line_number:
            return True
            break
        else:
            return False

def read_file(file_name="default"):
    #keep asking for input file, until file successully load
    while True:
        if file_name == "default":
            #ask for file name, if not given by the function
            file_name = input("Enter a file name: ")
            print()
        #try open file
        try:    
            txt = open(file_name, "r")
        #print error if open file failed, ask for new input
        except:
            print("Error -- ", end="")
            file_name = input("Enter a file name: ")
            print()
        #break loop if no error is detected
        else:
            break
    return txt

def read_data(txt):
    #track line number, and create an empty dictionary
    line_number = 0
    word_dict = {}
    #read file line by line
    for line in txt:
        line_number += 1
        #remove txt format 
        line = remove_format(line)
        #split the line word by word
        words = line.split()

        for word in words:            
            #remove none alphabetic words
            if not word.isalpha():
                words.remove(word)
            #exclude word with less than 2 characters
            if len(word) > 1:
            #check if word exist in dictionary
                #append line number to exist words
                if word in word_dict:
                    if check_exist_in_dict(word_dict, word, line_number) == False:
                        word_dict[word].append(line_number)
                #create new key for new word
                else:
                    word_dict[word] = [line_number]
    #sort the dictionary into ascending alphabatic orders
    sort_word_dict = dict(sorted(word_dict.items()))
    return sort_word_dict
        
def find_cooccurance(word_dict, input_str):
    #only runs legal input
    try:
        #split the input string in case of multipe word input
        input_list = input_str.split()
        #create an empty list to store line number
        line_number_list = []
        #find line number for each word
        for word in input_list:
            #check if word in dictionary, return None if not in dictionary
            if word not in word_dict:
                return "None"
            #append corrseponding line number
            line_number_list.append(word_dict[word])

        #find intersection of the multiple number list
        cooc = line_number_list[0]
        for number in line_number_list:
            cooc = list(set(cooc) & set(number))
        #if there is cooccurance, return line number list
        if len(cooc) > 0:
            cooc.sort()
            return cooc
        #else return None
        else:
            return "None"
    #return None if input is illegal
    except:
        return "None"

            
def display_result(cooc_list, input_str):
    print("The co-occurance for: ", end="")
    input_str = input_str.lower()
    input_list = input_str.split()
    for num in range(len(input_list)):
        if num != len(input_list)-1:
            print(input_list[num], end=", ")
        else:
            print(input_list[num], end="")
    print()
    print("Lines: ", end="")
    if cooc_list == "None":
        print("None")
    else:
        for index in range(len(cooc_list)):
            if index == len(cooc_list)-1:
                print(cooc_list[index], end="\n")
            else:
                print(cooc_list[index], end=", ")
    print()
        
def main():
    txt = read_file()
    word_dict = read_data(txt)
    while True:
        input_str = input("Enter space-separated words:")
        clean_input_str = remove_format(input_str)
        if clean_input_str == "q":
            break
        cooc_list = find_cooccurance(word_dict, clean_input_str)
        display_result(cooc_list, input_str)        
```

```python
if __name__ == "__main__":
    main()
```
